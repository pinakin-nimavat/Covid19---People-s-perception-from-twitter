```python
df_chg_0_lock = pd.read_csv(r"D:\Temp\As\change.org_topic_0_lockdowns.csv")

score_0_96_53_lock = cohen_kappa_score(df_chg_0_lock.annotation_96, df_chg_0_lock.annotation_53)
score_0_96_93_lock = cohen_kappa_score(df_chg_0_lock.annotation_96,df_chg_0_lock.annotation_93)
score_0_96_94_lock = cohen_kappa_score(df_chg_0_lock.annotation_96,df_chg_0_lock.annotation_94)
score_0_96_95_lock = cohen_kappa_score(df_chg_0_lock.annotation_96,df_chg_0_lock.annotation_95)
score_0_53_93_lock = cohen_kappa_score(df_chg_0_lock.annotation_53,df_chg_0_lock.annotation_93)
score_0_53_94_lock = cohen_kappa_score(df_chg_0_lock.annotation_53,df_chg_0_lock.annotation_94)
score_0_53_95_lock = cohen_kappa_score(df_chg_0_lock.annotation_53,df_chg_0_lock.annotation_95)
score_0_93_94_lock = cohen_kappa_score(df_chg_0_lock.annotation_93,df_chg_0_lock.annotation_94)
score_0_93_95_lock = cohen_kappa_score(df_chg_0_lock.annotation_93,df_chg_0_lock.annotation_95)
score_0_94_95_lock = cohen_kappa_score(df_chg_0_lock.annotation_94,df_chg_0_lock.annotation_95)

avg_score_0_96_lock = (score_0_96_53_lock + score_0_96_93_lock + score_0_96_94_lock + score_0_96_95_lock)/4
avg_score_0_53_lock = (score_0_96_53_lock + score_0_53_93_lock + score_0_53_95_lock + score_0_93_94_lock)/4
avg_score_0_93_lock = (score_0_96_93_lock + score_0_53_93_lock + score_0_93_94_lock + score_0_93_95_lock)/4
avg_score_0_94_lock = (score_0_96_94_lock + score_0_53_94_lock + score_0_93_94_lock + score_0_94_95_lock)/4
avg_score_0_95_lock = (score_0_96_95_lock + score_0_53_95_lock + score_0_93_95_lock + score_0_94_95_lock)/4

print("Average Score of Each Annotator for Lockdown File \n")
print("Average Score of Annotator 96:",avg_score_0_96_lock)
print("Average Score of Annotator 53:",avg_score_0_53_lock)
print("Average Score of Annotator 93:",avg_score_0_93_lock)
print("Average Score of Annotator 94:",avg_score_0_94_lock)
print("Average Score of Annotator 95:",avg_score_0_95_lock)

```

Since the average score of Annotator 53 is less than 0.2 we are eliminating Annotator_53 decision

```python
df_chg_0_lock = df_chg_0_lock.drop(columns='annotation_53',axis = 1)
df_chg_0_lock
```

```python
#Calculating Mode of Entire DataFrame
mode_0_lock = df_chg_0_lock.mode(axis = 1)

#Creating DataFrame so that we can Merge
df_mode_0_lock = pd.DataFrame(data = mode_0_lock)
df_mode_0_lock
```

We can see that we have 2 columns for mode as there were equal values for TRUE and FALSE in some rows.
Now we need to take highest AVG Kappa Score of Annotator for such rows.

From above we know that Annotator 94 has the highest Avg score.

```python
#Merging
new = [df_chg_0_lock,df_mode_0_lock]
new_df = pd.concat(new,axis = 1)

#Renaming the column name 
new_df = new_df.rename(columns={0:"Final_Value" , 1:"Value2"})
new_df
```

Setting the Annotators 94 Value to Final Value Column
