# 1.1) Implementation for twitter_topic_0_lockdown CSV file

```python
df_twitter_0_lockdown = pd.read_csv("twitter_topic_0_lockdowns.csv")

score_0_104_101_lockdown = cohen_kappa_score(df_twitter_0_lockdown.annotation_104, 
                                             df_twitter_0_lockdown.annotation_101)
score_0_104_102_lockdown = cohen_kappa_score(df_twitter_0_lockdown.annotation_104, 
                                             df_twitter_0_lockdown.annotation_102)
score_0_104_103_lockdown = cohen_kappa_score(df_twitter_0_lockdown.annotation_104, 
                                             df_twitter_0_lockdown.annotation_103)
score_0_101_102_lockdown = cohen_kappa_score(df_twitter_0_lockdown.annotation_101, 
                                             df_twitter_0_lockdown.annotation_102)
score_0_101_103_lockdown = cohen_kappa_score(df_twitter_0_lockdown.annotation_101, 
                                             df_twitter_0_lockdown.annotation_103)
score_0_102_103_lockdown = cohen_kappa_score(df_twitter_0_lockdown.annotation_102, 
                                             df_twitter_0_lockdown.annotation_103)
```

```python
avg_score_0_104_lockdown = (score_0_104_101_lockdown + score_0_104_102_lockdown + score_0_104_103_lockdown)/3
avg_score_0_101_lockdown = (score_0_104_101_lockdown + score_0_101_102_lockdown + score_0_101_103_lockdown)/3
avg_score_0_102_lockdown = (score_0_104_102_lockdown + score_0_101_102_lockdown + score_0_102_103_lockdown)/3
avg_score_0_103_lockdown = (score_0_104_103_lockdown + score_0_101_103_lockdown + score_0_102_103_lockdown)/3
```

```python
print("Average Score of Each Annotator for Lockdown File \n")
print("Average Score of Annotator 104:",avg_score_0_104_lockdown)
print("Average Score of Annotator 101:",avg_score_0_101_lockdown)
print("Average Score of Annotator 102:",avg_score_0_102_lockdown)
print("Average Score of Annotator 103:",avg_score_0_103_lockdown)
```

```python
list = [avg_score_0_104_lockdown,avg_score_0_101_lockdown,avg_score_0_102_lockdown,avg_score_0_103_lockdown]

for i in list:
    if i>0.2:
        print (i)
```

Since We can see that the avergae kappa score for annotator 104 and 102 is less than 0.2 we will drop them.

```python
df_twitter_0_lockdown = df_twitter_0_lockdown.drop(columns=['annotation_104', 'annotation_102'])
df_twitter_0_lockdown
```

```python
#Calculating Mode of Entire DataFrame
mode_0_lock = df_twitter_0_lockdown.mode(axis = 1)

#Creating DataFrame so that we can Merge
df_mode_0_lock = pd.DataFrame(data = mode_0_lock)
df_mode_0_lock
```

We can see that we have 3 columns for mode as there were equal values for TRUE and FALSE in some rows. Now we need to take highest AVG Kappa Score of Annotator for such rows.

From above we know that Annotator 103 has the highest Avg score.

```python
#Merging
new01 = [df_twitter_0_lockdown,df_mode_0_lock]
new_df01 = pd.concat(new01,axis = 1)

#Renaming the column name 
new_df01 = new_df01.rename(columns={0:"Final_Value" , 1:"Value2", 2:"Value3"})
new_df01
```

```python
for idx,rows in new_df01.iterrows():
    if (new_df01.Value2[idx] or new_df01.Value3[idx] == True) or (new_df01.Value2[idx] or 
                                                                  new_df01.Value3[idx]) == False:
        new_df01.Final_Value[idx] = new_df01.annotation_103[idx]
```

Setting the annotator 103 value to final value column 

```python
final_df_0_lock = new_df01.drop(['Value2',"Value3"],axis = 1)
final_df_0_lock
```
